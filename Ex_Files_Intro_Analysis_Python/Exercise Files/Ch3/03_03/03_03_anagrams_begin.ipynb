```python
wordclean = sorted(list(set([word.strip().lower() for word in open('words','r')])))
```

```python
wordclean[:10]
```

```python
sorted('elvis')
```

```python
sorted('elvis')==sorted('lives')
```

```python
sorted('love')==sorted('hate')
```

```python
def signature(word):
    return ''.join(sorted(word))
    
```

```python
signature('elvis')
```

```python
def anagram(myword):
    return [word for word in wordclean if signature(word) == signature(myword)]
```

```python
anagram('dictionary')
```

```python
%timeit anagram('dictionary')
```

```python
word_bysig = {}

for word in wordclean:
    word_bysig[signature(word)].append(word)
```

```python
import collections
```

```python
word_bysig = collections.defaultdict(list)
for word in wordclean:
    word_bysig[signature(word)].append(word)
```

```python
word_bysig
```

```python
def anagram_fast(myword):
    return word_bysig[signature(myword)]
```

```python
anagram_fast('dictionary')
```

```python
anagram_fast('a')
```

```python
anagrams_all = {word: anagram_fast(word) for word in wordclean if len(anagram_fast(word)) > 1}
```

```python
len(anagram_all)
```

```python
%timeit anagrams_all = {word: anagram_fast(word) for word in wordclean if len(anagram_fast(word)) > 1}
```
